```python
world_happiness_2019_df = pd.read_csv(os.path.join("2019whr.csv"))
```

```python
# Will be checking if theres any correlation between these factors to the score
# by using a scatterplot
```

```python
# This will be GDP per Capita
gdp_x = world_happiness_2019_df.Score.values.reshape(-1,1)
gdp_y = world_happiness_2019_df.GDP_per_capita.values.reshape(-1,1)
gdp_plot = plt.scatter(gdp_x, gdp_y)
```

```python
# This graph will be for Social support
#(having friends, family to turn to in times of need)

ss_x = world_happiness_2019_df.Score.values.reshape(-1,1)
ss_y = world_happiness_2019_df.Social_support.values.reshape(-1,1)
ss_plot = plt.scatter(ss_x, ss_y)
```

```python
# This is for freedom to make life choices 
freedom_x = world_happiness_2019_df.Score.values.reshape(-1,1)
freedom_y = world_happiness_2019_df.Freedom.values.reshape(-1,1)
health_plot = plt.scatter(freedom_x, freedom_y)
```

```python
# This is for Generosity 
genorosity_x = world_happiness_2019_df.Score.values.reshape(-1,1)
genorosity_y = world_happiness_2019_df.Generosity.values.reshape(-1,1)
genorosity_plot = plt.scatter(genorosity_x, genorosity_y)
```
