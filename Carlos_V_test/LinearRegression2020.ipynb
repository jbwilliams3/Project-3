```python
# MUST pip install "pip install -U scikit-learn scipy matplotlib"
# to be able to use sklearn.also known as sklearn is a python library to implement
#machine learning models and statistical modelling.
#Through scikit-learn, we can implement various machine 
#learning models for regression, classification, clustering, and
#statistical tools for analyzing these models.

%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import os
```

```python
WH2020_df = pd.read_csv(os.path.join("2020-cleaned.csv"))
```

```python
WH2020_df.head()
```

```python
WH2020_df.columns
```

```python
# Renaming Columns here to so plotting is easier
WH2020_df.rename(columns={"Healthy life expectancy": "Health", "Freedom to make life choices": "Freedom", "Perceptions of corruption": "Corruption"}, inplace=True)
WH2020_df.columns
```

```python
# This will be GDP per Capita
gdp_x = WH2020_df.HappinessScore.values.reshape(-1,1)
gdp_y = WH2020_df.GDP_per_capita.values.reshape(-1,1)
gdp_plot = plt.scatter(gdp_x, gdp_y)
plt.xlabel("Happiness Score")
plt.ylabel("GDP Score")
```

```python
# This will be the Social Support graph
ss_x = WH2020_df.HappinessScore.values.reshape(-1,1)
ss_y = WH2020_df.SocialSupport.values.reshape(-1,1)
gdp_plot = plt.scatter(ss_x, ss_y)
plt.xlabel("Happiness Score")
plt.ylabel("Social Support Score")
```

```python
# This will be the Health graph
health_x = WH2020_df.HappinessScore.values.reshape(-1,1)
health_y = WH2020_df.Health.values.reshape(-1,1)
health_plot = plt.scatter(health_x, health_y)
plt.xlabel("Happiness Score")
plt.ylabel("Health Score")
```

```python
# This will be the Freedom graph
freedom_x = WH2020_df.HappinessScore.values.reshape(-1,1)
freedom_y = WH2020_df.Freedom.values.reshape(-1,1)
freedom_plot = plt.scatter(freedom_x, freedom_y)
plt.xlabel("Happiness Score")
plt.ylabel("Freedom Score")
```

```python
# This will be the Generosity graph
generosity_x = WH2020_df.HappinessScore.values.reshape(-1,1)
generosity_y = WH2020_df.Generosity.values.reshape(-1,1)
generosity_plot = plt.scatter(generosity_x, generosity_y)
plt.xlabel("Happiness Score")
plt.ylabel("Generosity Score")
```

```python
# This will be the Perception of corruption graph
corruption_x = WH2020_df.HappinessScore.values.reshape(-1,1)
corruption_y = WH2020_df.Corruption.values.reshape(-1,1)
corruption_plot = plt.scatter(corruption_x, corruption_y)
plt.xlabel("Happiness Score")
plt.ylabel("Corruption Score")
```

```python
# Creating linear regresion models, fit the models, assign variables to hold coefficients and Y intercepts
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

###The train_test_split() method is used to split our data into train and test sets.
###First, we need to divide our data into features (X) and labels (y).
###The dataframe gets divided into X_train, X_test, y_train, and y_test.
###X_train and y_train sets are used for training and fitting the model

gdpX_train, gdpX_test, gdpY_train, gdpY_test = train_test_split(gdp_x, gdp_y, random_state = 1)
ssX_train, ssX_test, ssY_train, ssY_test = train_test_split(ss_x, ss_y, random_state = 1)
healthX_train, healthX_test, healthY_train, healthY_test = train_test_split(health_x, health_y, random_state = 1)
freedomX_train, freedomX_test, freedomY_train, freedomY_test = train_test_split(freedom_x, freedom_y, random_state = 1)
generosityX_train, generosityX_test, generosityY_train, generosityY_test = train_test_split(generosity_x, generosity_y, random_state = 1)
corruptionX_train, corruptionX_test, corruptionY_train, corruptionY_test = train_test_split(corruption_x, corruption_y, random_state = 1)
```

```python
GDPmodel = LinearRegression()
GDPmodel.fit(gdpY_train, gdpX_train)
print(GDPmodel.score(gdpY_test, gdpX_test))

y_min = gdp_y.min()
y_max = gdp_y.max()

Predicted_GDP_x_min = GDPmodel.predict([[y_min]])
Predicted_GDP_x_max = GDPmodel.predict([[y_max]])

plt.scatter(gdp_x, gdp_y)
plt.plot([Predicted_GDP_x_min[0], Predicted_GDP_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("GDP Score")
```

```python
Socialmodel = LinearRegression()
Socialmodel.fit(ssY_train, ssX_train)
print(Socialmodel.score(ssY_test, ssX_test))

y_min = ss_y.min()
y_max = ss_y.max()

Predicted_Social_x_min = Socialmodel.predict([[y_min]])
Predicted_Social_x_max = Socialmodel.predict([[y_max]])

plt.scatter(ss_x, ss_y)
plt.plot([Predicted_Social_x_min[0], Predicted_Social_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("Social Support Score")
```

```python
Healthmodel = LinearRegression()
Healthmodel.fit(healthY_train, healthX_train)
print(Healthmodel.score(healthY_test, healthX_test))

y_min = health_y.min()
y_max = health_y.max()

Predicted_Health_x_min = Healthmodel.predict([[y_min]])
Predicted_Health_x_max = Healthmodel.predict([[y_max]])

plt.scatter(health_x, health_y)
plt.plot([Predicted_Health_x_min[0], Predicted_Health_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("Health Score")
```

```python
Freedom_model = LinearRegression()
Freedom_model.fit(freedomY_train, freedomX_train)
print(Freedom_model.score(freedomY_test, freedomX_test))

y_min = freedom_y.min()
y_max = freedom_y.max()

Predicted_Freedom_x_min = Freedom_model.predict([[y_min]])
Predicted_Freedom_x_max = Freedom_model.predict([[y_max]])

plt.scatter(freedom_x, freedom_y)
plt.plot([Predicted_Freedom_x_min[0], Predicted_Freedom_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("Freedom Score")
```

```python
Generosity_model = LinearRegression()
Generosity_model.fit(generosityY_train, generosityX_train)
print(Generosity_model.score(generosityY_test, generosityX_test))

y_min = generosity_y.min()
y_max = generosity_y.max()

Predicted_Generosity_x_min = Generosity_model.predict([[y_min]])
Predicted_Generosity_x_max = Generosity_model.predict([[y_max]])

plt.scatter(generosity_x, generosity_y)
plt.plot([Predicted_Generosity_x_min[0], Predicted_Generosity_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("Generosity Score")
```

```python
Corruption_model = LinearRegression()
Corruption_model.fit(corruptionY_train, corruptionX_train)
print(Corruption_model.score(corruptionY_test, corruptionX_test))

y_min = corruption_y.min()
y_max = corruption_y.max()

Predicted_Corruption_x_min = Corruption_model.predict([[y_min]])
Predicted_Corruption_x_max = Corruption_model.predict([[y_max]])

plt.scatter(corruption_x, corruption_y)
plt.plot([Predicted_Corruption_x_min[0], Predicted_Corruption_x_max[0]], [y_min, y_max], color = "red")
plt.xlabel("Happiness Score")
plt.ylabel("Corruption Score")
```

```python
WH2020_df["GDP_X"]=gdp_x
WH2020_df["GDP_y"]=gdp_y
WH2020_df["Health_X"]=health_x
WH2020_df["Health_y"]=health_y
WH2020_df["Social_X"]=ss_x
WH2020_df["Social_y"]=ss_y
WH2020_df["Generosity_X"]=generosity_x
WH2020_df["Generosity_y"]=generosity_y
WH2020_df["Freedom_X"]=freedom_x
WH2020_df["Freedom_y"]=freedom_y
WH2020_df["Corruption_X"]=corruption_x
WH2020_df["Corruption_y"]=corruption_y


WH2020_df.to_csv("2020_predicted.csv", index=False)
```
