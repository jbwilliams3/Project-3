```python
country_location_df["Lat"] = ""
country_location_df["Lng"] = ""
country_location_df.head()
```

```python
for index, row in country_location_df.iterrows():
    city = row['country']
    results = geocode_results(f"{city}")

    if results:  # Check if results is not empty
        if len(results) > 0:  # Check if results has elements
            country_location_df.loc[index, "Lat"] = results[0]["geometry"]["location"]["lat"]
            country_location_df.loc[index, "Lng"] = results[0]["geometry"]["location"]["lng"]
        else:
            print(f"No results found for {city}.")
    else:
        print(f"No results found for {city}.")
```
