```python
country_list = []
def list_country(l):
    for country in l:
        if country in country_list:
            continue
        else:
            country_list.append(country)
```

```python
list_country(list_2019)
list_country(list_2020)
list_country(list_2021)

country_list.sort()
```

```python
country_list
```

```python
country_id = []
counter = 1
for country in country_list:
    country_id.append(counter)
    counter += 1
```

```python
country_data = {'country':country_list, 'country_id':country_id}

country_data
```

```python
country_df = pd.DataFrame.from_dict(country_data)

country_df = pd.read_csv(os.path.join(r'C:\Users\roelj\OneDrive\Desktop\school\Project-3\csv\countries.csv'))
```

```python
country_id_dict = dict()
for index,row in country_df.iterrows() :
    country_name = row['country']
    country_id = row['country_id']
    country_id_dict[country_name] = country_id
```

```python
country_id_dict
```

```python
happiness_19_df['country_id'] =""
happiness_20_df['country_id'] =""
happiness_21_df['country_id'] =""
```

```python
def add_country_id(df):
    for index,row in df.iterrows() :
        try:
            df.loc[index,'country_id'] = country_id_dict.get(row['Country'])
        except:
            continue
```

```python
add_country_id(happiness_19_df)
add_country_id(happiness_20_df)
add_country_id(happiness_21_df)
```
